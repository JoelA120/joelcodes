```python
af.groupby('make')['price'].min().reset_index()
```

```python
af.groupby('make')['price'].std().min()
```

```python
af.groupby(['make','fuel_type']).mean().tail(2)
```

```python
af_group = af.groupby(['make','fuel_type'])
af_agg_price = af_group['price'].agg(['max','min','mean'])
print(af_agg_price)
```

```python
dic_1 = {"make":af.make, "price":af.price, "fuel_type":af.fuel_type}
af_1 = pd.DataFrame(dic_1)
print(af_1)
```

```python
dic_2 = {"num_of_doors":af.num_of_doors, "engine_size":af.engine_size, "wheel_base":af.wheel_base}
af_2 = pd.DataFrame(dic_2)
print(af_2)
```

```python
merged_dics = pd.merge(af_1,af_2,on=af.make)
print(merged_dics)
```
